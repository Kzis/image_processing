```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./picme.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img_gray = cv2.imread('./picme.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')
```

```python
def colorReduce(inimg, scale=30):
    return (inimg/scale).astype(np.uint8)*scale
```

```python
plt.imshow(cv2.cvtColor(colorReduce(img), cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(colorReduce(img_gray), cmap='gray')
```
